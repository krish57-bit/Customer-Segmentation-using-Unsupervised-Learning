Feature Extraction

The following code creates features that would be helpful to describe the profile of the customer

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.preprocessing import StandardScaler

# Creating a copy of the raw dataframe
raw_data = data.copy()  # saving the raw dataframe for reference

# Creating a new field to store the Age of the customer
data['Age'] = 2022 - data['Year_Birth']

# Recoding the customer's education level to numeric form
data["Education"].replace({"Basic": 0, "2n Cycle": 1, "Graduation": 2, "Master": 3, "PhD": 4}, inplace=True)

# Recoding the customer's marital status to numeric form
data['Marital_Status'].replace({"Married": 1, "Together": 1, "Absurd": 0, "Widow": 0, "YOLO": 0, "Divorced": 0,
                                "Single": 0, "Alone": 0}, inplace=True)

# Creating a new field to store the number of children in the household
data['Children'] = data['Kidhome'] + data['Teenhome']

# Creating Family_Size
data['Family_Size'] = data['Marital_Status'] + data['Children'] + 1

# Creating a new field to store the total spending of the customer
data['Total_Spending'] = data["MntWines"] + data["MntFruits"] + data["MntMeatProducts"] + data["MntFishProducts"] + data["MntSweetProducts"] + data["MntGoldProds"]
data["Total_Promo"] = data["AcceptedCmp1"] + data["AcceptedCmp2"] + data["AcceptedCmp3"] + data["AcceptedCmp4"] + data["AcceptedCmp5"]

# Calculating how long the customer has been with the company
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'])
today = datetime.today()
data['Days_as_Customer'] = (today - data['Dt_Customer']).dt.days

# Calculating the total number of promotions the customer responded to
data['Offers_Responded_To'] = data['AcceptedCmp1'] + data['AcceptedCmp2'] + data['AcceptedCmp3'] + data['AcceptedCmp4'] + data['AcceptedCmp5'] + data['Response']

# Creating Parental Status
data["Parental_Status"] = np.where(data["Children"] > 0, 1, 0)

# Dropping columns which are already used to create new features
columns_to_drop = ['Year_Birth', 'Kidhome', 'Teenhome']
data.drop(columns=columns_to_drop, axis=1, inplace=True)

# Renaming columns for clarity
data.rename(columns={
    "Marital_Status": "Marital Status",
    "MntWines": "Wines",
    "MntFruits": "Fruits",
    "MntMeatProducts": "Meat",
    "MntFishProducts": "Fish",
    "MntSweetProducts": "Sweets",
    "MntGoldProds": "Gold",
    "NumWebPurchases": "Web",
    "NumCatalogPurchases": "Catalog",
    "NumStorePurchases": "Store",
    "NumDealsPurchases": "Discount Purchases"
}, inplace=True)

# Selecting relevant columns
data = data[["Age", "Education", "Marital Status", "Parental_Status", "Children", "Income", "Total_Spending", "Days_as_Customer",
             "Recency", "Wines", "Fruits", "Meat", "Fish", "Sweets", "Gold", "Web", "Catalog", "Store", "Discount Purchases",
             "Total_Promo", "NumWebVisitsMonth"]]

# Handling missing values by filling with the mean (numeric columns only)
data.fillna(data.mean(), inplace=True)

# Scaling numerical features
scaler = StandardScaler()
scaled_columns = ['Income', 'Total_Spending', 'Days_as_Customer', 'Total_Promo', 'NumWebVisitsMonth']
data[scaled_columns] = scaler.fit_transform(data[scaled_columns])

# Now data is ready for further analysis or modeling
```

```python
columns = [column for column in data.columns if column != "ID"]

numeric_features = [feature for feature in columns if data[feature].dtype != 'O']

continuous_features=[feature for feature in numeric_features if len(data[feature].unique())>=10]
print('Names of continues features :',continuous_features)
```

### Multicollinearity Check
#### Variance Inflation Factor (VIF)
* Multicollinearity occurs when there are two or more independent variables in a multiple regression model, which have a high correlation among themselves. When some features are highly correlated.
* Multicollinearity can be detected using various techniques, one such technique being the Variance Inflation Factor(VIF).

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def compute_vif(considered_features, data):
    X = data[considered_features]
    # The calculation of variance inflation requires a constant
    X['intercept'] = 1

    # Create dataframe to store VIF values
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable'] != 'intercept']
    return vif

# We will remove the datetime column which is not required for the VIF checking
cont_features = continuous_features.copy()

# We will not check VIF for Total_Spending, as it is the summation of Wines, Fruits, Meat, Fish, Sweets, Gold.
cont_features.remove('Total_Spending')

compute_vif(cont_features, data)
```
